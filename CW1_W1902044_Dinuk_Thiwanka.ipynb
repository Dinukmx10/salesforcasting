# 3. Market Segmentation Using Clustering Algorithms

## I. Introduction 

Market segmentation is the basis of marketing, helping divide target markets into separate units based on similarities. It allows companies to develop personalized strategies for different segments, improving customer interactions. Clustering algorithms are key in this process, as they help create meaningful patterns from data.

## II. Clustering Algorithms in Market Segmentation

Clustering algorithms are a form of unsupervised machine learning. They don't use labeled data but instead find patterns within the data and form natural groups. Some of the best clustering algorithms for market segmentation include:

- **K-Means**: A popular clustering method that splits data into a user-defined number of clusters to minimize variance. It works well for customer segmentation based on numerical attributes like age, income, or expenditure.
  
- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: Enhances data arrangement by clustering based on the density of surrounding data points. It's particularly useful for noisy datasets and customer segmentation based on behavioral information.

## III. Applications and Benefits of Clustering in Market Segmentation

Clustering plays a vital role in market segmentation by enabling businesses to create customer-oriented strategies. Some benefits include:

- **Personalized Advertising**: Segmenting customers based on behavior, preferences, and demographics allows for targeted marketing strategies. For example, a fashion retailer might target young customers with casual attire ads and professionals with formal outfits.
  
- **Product Recommendation Systems**: Clustering enhances recommendation systems, helping e-commerce platforms suggest products based on purchasing behavior.
  
- **Customer Retention**: Clustering helps identify at-risk customers, enabling businesses to implement re-engagement plans.
  
- **Improved Targeting Accuracy**: It leads to cost-efficient marketing by focusing resources on high-potential segments and gaining deeper insights into customer needs.

## IV. Conclusion

Clustering algorithms are key to market segmentation, helping businesses identify the right customer groups. K-Means and DBSCAN are popular models for clustering, and the future holds numerous opportunities for applying clustering techniques in marketing.

```python



# Define the SMAPE function 
def smape(y_true, y_pred):
    
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    denominator = np.abs(y_true) + np.abs(y_pred)
    numerator = np.abs(y_true - y_pred)
    smape_values = np.where(denominator == 0, 0.0, numerator / denominator)
    return 200 * np.mean(smape_values)


if train_new[col].shape[0] != y_target.shape[0]:
    print(f"Adjusting y_target to match feature rows: {train_new[col].shape[0]} rows.")
    y_target = y_target[:train_new[col].shape[0]]

# Convert data to DMatrix for XGBoost
train_dmatrix = xgb.DMatrix(train_new[col], label=y_target)
test_dmatrix = xgb.DMatrix(X_test[col], label=y_test)


params = {
    'objective': 'reg:squarederror', 
    'eval_metric': 'rmse',           
    'eta': 0.05,                      
    'max_depth': 6,                   
    'n_estimators': 100,              
    'learning_rate': 0.01,           
    'subsample': 0.8,                 
    'colsample_bytree': 0.8,          
    'alpha': 0.1,                     
    'lambda': 0.1                     
}

# Custom SMAPE evaluation function
def smape_eval(y_pred, dtrain):
    y_true = dtrain.get_label()
    return 'smape', smape(y_true, y_pred)

# List to store trained models
model_sets = []

# Train the XGBoost model
print("Training XGBoost model...")
model = xgb.train(
    params=params,
    dtrain=train_dmatrix,
    num_boost_round=100,
    evals=[(train_dmatrix, 'train'), (test_dmatrix, 'test')],
    feval=smape_eval,  
    early_stopping_rounds=20, 
    verbose_eval=10  
)

# Store the trained model
model_sets.append(model)

# Predictions
y_train_pred_xgb = model.predict(train_dmatrix)
y_test_pred_xgb = model.predict(test_dmatrix)

# Evaluation Metrics
train_smape = smape(y_target, y_train_pred_xgb)
test_smape = smape(y_test, y_test_pred_xgb)

print(f"SMAPE error - Train: {train_smape:.4f}, Test: {test_smape:.4f}")

# Log-transformed Mean Absolute Error
train_mae = mean_absolute_error(np.log1p(y_target), np.log1p(y_train_pred_xgb))
test_mae = mean_absolute_error(np.log1p(y_test), np.log1p(y_test_pred_xgb))

print(f"Log-Transformed MAE - Train: {train_mae:.4f}, Test: {test_mae:.4f}")


```
